```python
import numpy as np
from scipy.spatial import ConvexHull

def encircle(x0,y0, ax=None, **kw):
    if not ax: ax=plt.gca()
    p = np.c_[x0,y0]
    hull = ConvexHull(p)
    poly = plt.Polygon(p[hull.vertices,:], **kw)
    ax.add_patch(poly)

def encircle(x2,y2, ax=None, **kw):
    if not ax: ax=plt.gca()
    p = np.c_[x2,y2]
    hull = ConvexHull(p)
    poly = plt.Polygon(p[hull.vertices,:], **kw)
    ax.add_patch(poly)


plt.scatter( x0, y0, label= "window open" )
plt.scatter( x2, y2, label= "window closed", color="red" )

plt.title("Temperature - Window State Graph")
plt.xlabel("Outdoor Temperature")
plt.ylabel("Indoor Temperature")

encircle(x0, y0, alpha=0.1)
encircle(x2, y2, alpha=0.1, color= "red")

plt.scatter(sum(x0)/len(x0), sum(y0)/len(y0), color = "black")
plt.scatter(sum(x2)/len(x2), sum(y2)/len(y2), color = "black")

#, ec=”k”,  fc=”purple”, 
plt.legend()
plt.show()
```
